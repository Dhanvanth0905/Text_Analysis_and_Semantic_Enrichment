```python
import pandas as pd
url = "https://raw.githubusercontent.com/Jayasurya-Marasani/datasets/main/booking_reviews.csv"
df = pd.read_csv(url)
df.head()
```

```python
df = df[['hotel_name','review_text']]
df = df.drop_duplicates()
d = {}
for index, row in df.iterrows():
    if row['hotel_name'] in d:
        d[row['hotel_name']] += '. ' + row['review_text']
    else:
        d[row['hotel_name']] = row['review_text']
hotel_reviews = d
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

```python
counts = {}
for hotel, review in hotel_reviews.items():
    words = word_tokenize(review)
    words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]
    words = [lemmatizer.lemmatize(word) for word in words]
    pos_tags = nltk.pos_tag(words)
    for word, pos in pos_tags:
        if pos.startswith("JJ"):
            counts[word] = counts.get(word, 0) + 1
    descriptors = sorted(counts.items(), key=lambda x: x[1], reverse=True)[:5]
    print(hotel)
    for descriptor, count in descriptors:
        print(f"{descriptor}: {count}")
```

Regex Tokenizer

```python
from nltk.tokenize import RegexpTokenizer
for i in range(100):
  text=df.review_text[i]
  num_tokenizer = RegexpTokenizer(r'\d+\.?\d*')
  numbers = num_tokenizer.tokenize(text)
  alpha_tokenizer = RegexpTokenizer(r'[a-zA-Z]+')
  letters = alpha_tokenizer.tokenize(text)
  print("Numbers:", numbers)
  print("Letters:", letters)
```

Lesk Algorithm

```python
import nltk
from nltk.wsd import lesk

def get_semantic(seq, key_word):
    temp = word_tokenize(seq)
    temp = lesk(temp, key_word)
    return temp.definition()
```

```python
for i in range(100):
  text=df.review_text[i]
  keyword='good'
  print(get_semantic(text, keyword))
```
